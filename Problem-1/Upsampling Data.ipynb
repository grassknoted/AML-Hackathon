### Upsample images of NOT female

```python
transformations = ['contrast', 'h_flip']

for index, row in not_female.iterrows():
    im = cv2.imread('../../Selfie-dataset/images/'+str(row['image_name'])+'.jpg')
    print(row['image_name'])
    for transform in transformations:
        if(transform == 'contrast'):
            array_alpha = np.array([1.25])
            array_beta = np.array([-100.0])
            cv2.add(im, array_beta, im)                    
            cv2.multiply(im, array_alpha, im)
            cv2.imwrite( str(row['image_name'])+'_c.jpg', im)
        elif(transform == 'h_flip'):
            im = cv2.flip(im, 1)
            cv2.imwrite( str(row['image_name'])+'_h.jpg', im)
#         elif(transform == 'gamma'):
#             img = cv2.imread('../../Hackathon/Selfie-dataset/images/'+str(row['image_name'])+'.jpg')
#             invGamma = 1.0 / 0.9
#             table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)])
#             im = cv2.LUT(img.astype(np.uint8), table.astype(np.uint8))
#             cv2.imwrite("../../Upsampled_Images/"+str(row['image_name'])+'_g.jpg', im);
            
```

## Upsampling baby, child, and senior

```python
transformations = ['contrast', 'h_flip']
row_data = []
for index, row in baby.iterrows():
    im = cv2.imread('../../Selfie-dataset/images/'+str(row['image_name'])+'.jpg')
    
    for transform in transformations:
        if(transform == 'contrast'):
            array_alpha = np.array([1.25])
            array_beta = np.array([-100.0])
            cv2.add(im, array_beta, im)                    
            cv2.multiply(im, array_alpha, im)
            for i in range(0, 2):
                temp = row.copy()
                cv2.imwrite('../../Child/'+str(row['image_name'])+'_c'+str(i)+'.jpg', im)
                temp[0] = str(row['image_name'])+'_c'+str(i)
                row_data.append(temp)
        
        elif(transform == 'h_flip'):
            im = cv2.flip(im, 1)
            for i in range(0, 2):
                temp2 = row.copy()
                cv2.imwrite('../../Child/'+str(row['image_name'])+'_h'+str(i)+'.jpg', im)
                temp2[0] = str(row['image_name'])+'_h'+str(i)
                row_data.append(temp2)
```

```python
with open("child.csv", 'w') as f:
    writer = csv.writer(f)
    writer.writerows(row_data)
```
